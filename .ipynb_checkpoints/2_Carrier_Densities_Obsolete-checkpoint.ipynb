# Carrier Densities at Finite Temperature

The expressions derived in [Background on the Band Structure of Bilayer Graphene](1_Background_on_Bilayer_Graphene_BandStructure.ipynb) allow us to derive expression for both the total carrier density on the two layers, $n_+=n_1+n_2$, and the carrier density difference between the two layers.This notebook describes how to work with the Fermi-Dirac distribution and use it to compute the carrier density $n_+$ at arbitrary temperature. We also introduce the interalyer carrier density difference and $n_-=n_1-n_2$, where $n_{1(2)}$ is the carrier density on the top (bottom) layer of graphene. As with finite temperature densities, this can only be computed with an integral, so we include it here rather than the section on the band structure of bilayer graphene.

## Total Carrier Density $n_+=n_1+n_2$ at $T=0$

At temperature $T=0$, there is an analytic formula for $n_+$. There is no analytic formula for $n_-=n_1-n_2$ at $T=0$ K. Its computation will require numerical integration, so we will hold off on its discussion until the $T\neq 0$ K section.

At $T=0$, there are minimum and maximum occupied Fermi wavevectors $k_{F\pm}$. In general then, the Fermi surface will be an annulus. The carrier density (including spin degeneracy $g_s=2$ and valley degeneracy $g_v=2$) then given by, writing explicitly as a function of $\epsilon_F$ and $u$,
$$ n_+(\epsilon_F,u)= g_sg_v \frac{\pi(k_{F+}(\epsilon_F,u)^2-k_{F-}(\epsilon_F,u)^2)}{(2\pi)^2}= \frac{k_{F+}(\epsilon_F,u)^2-k_{F-}(\epsilon_F,u)^2}{\pi} $$

Recall that this is a piecewise function equal to

$$ n_+ =\frac{1}{\pi}\left\{ \begin{array}{} 0 & \epsilon_F<\epsilon^* \\ k_{F+}^2 - k_{F-}^2 & \epsilon^*<\epsilon_F<u/2 \\  k_{F+}^2 & u/2 < \epsilon_F\end{array} \right.$$

This function is provided in `BLG.BandStructure.nplusT0`.

```python
from BLG.BandStructure import *
nplusT0?
```

Plotting for various applied potentials, we can explicitly see the gap.

```python
vplus = np.linspace(-0.1,0.1,num=200)
vminus  = np.linspace(0,0.05,num=3)

densities = nplusT0(vplus,vminus[:,np.newaxis])

for n,vminus in zip(densities,vminus):
    plt.plot(vplus*1000,n*10**-4, label = 'vminus = {} meV'.format(round(vminus*1000)))
    
plt.title('Electron Density (m^-2) at Various Gaps (T=0)')
plt.ylabel('n_+ (cm^{-2})')
plt.xlabel('Fermi Level (meV)')
plt.legend();
```

## Carrier densities at Non-zero Temperatures

Computation of $n_+$ and $n_-$ at nonzero temperature requires numerical integration.

### Using the Fermi-Dirac Distribution

We plot the Fermi-Dirac distribution (available in `BLG.BandStructure.StatisticalDistributions`) for various temperatures. Use `numpy arrays` for faster computation.

```python
from BLG.StatisticalDistributions import *

temps = np.array([0,50,300])
energies = np.linspace(-1,1,num=1000)[:,np.newaxis]*eVtoJ

plt.plot(energies*np.ones(len(temps)),FermiDirac(energies,temps));
```

## Plotting the Fermi-Dirac Distribution with a Dispersion Relation

We'll want to integrate over the Fermi-Dirac distribution using a particular dispersion relation $\epsilon(k)$. For simplicity, lets use the quadratic approximation to the BLG dispersion.

```python
def E(k):
    '''Returns the low energy quadratic approximation for BLG dispersion relation.'''
    return (hbar*vF*k)**2 / (g1)
```

This may include a potential energy $eV_+$. In that case, the total energy we will insert into the Fermi-Dirac distribution is $E(k)+qV_+$ where $q=-|e|$ for carriers in the conduction band and $q=|e|$ for carriers in the valence band.

We integrate over ALL wavevectors in the first Brillouin zone, since they are all occupied, though some will end up being heavily suppressed by the Fermi-Dirac distribution. The magnitudes of wavevectors are therefore $0\leq k \leq 1/\sqrt{3}a$ where $\sqrt{3}a$ is the unit cell spacing.

```python
ks = np.linspace(0,1/(np.sqrt(3)*a),num=1000)

plt.plot(ks[:1000],FermiDirac(E(ks[:1000])-q*0.01,T=100))
plt.show()
```

In reality, the valence band will contribute. This means the distribution over all wavevectors is

$$f(\epsilon-|e|V_+)-f(\epsilon+|e|V_+)$$

```python
ks = np.linspace(0,1/(np.sqrt(3)*a),num=10000)

plt.plot(ks[:1000],FermiDirac(E(ks[:1000])-q*0.01,T=100)-FermiDirac(E(ks[:1000])+q*0.01,T=100))
plt.show()
```

## Charge Density $n_+=n_1+n_2$  for $T\neq 0$

In two dimensions, the integral we wish to perform to determine the carrier density is given by

$$ n_+ = \frac{g_v g_s}{(2\pi)^2} \int\int d^2k (f(\epsilon-|e|V_+)-f(\epsilon+|e|V_+)) =\frac{2}{\pi} \int_0^{1/\sqrt{3}a}dk\quad k (f(\epsilon-|e|V_+)-f(\epsilon+|e|V_+))$$

where $g_v=2$ and $g_s=2$ account for valley and spin degeneracy respectively and we've performed the radial integral. Keep in mind we are ignoring the higher energy bands. For these purposes, a function `nplus` that yields the total carrier density $n_+$ at arbitrary temperature.

```python
nplus?
```

Recall that `nplusT0`provides an analytic formula for $n_+$ at $T=0$. We compare them here. They are offset for clarity. We see that the agreement is very good.

```python
T = 0

vplus = np.linspace(0,0.2,num=100)
vminus = 0.1


n1 = nplusT0(vplus,vminus)

n2 = nplus(vplus,vminus,T) + 10**16

n3 = nplus(vplus,vminus,50) + 10**16
```

```python
plt.plot(vplus*1000,n1*10**-4,label='Analytic (T=0 K)', linestyle=':',linewidth=2)
plt.plot(vplus*1000,n2*10**-4,label='Numeric (T=0 K)',linestyle='--',linewidth=2)
plt.plot(vplus*1000,n3*10**-4,label='Numeric (T=50 K)',linestyle='--',linewidth=2)
plt.ylabel('n_+ (cm^{-2})')
plt.xlabel('Fermi Level (meV)')
plt.title('Total Carrier Density (Numeric offset for clarity)')
plt.legend()
plt.show()
```

## Interlayer Charge Density $n_-=n_1-n_2$  for $T\neq 0$

If we wish to find the total carrier density in the top layer, we integrate

$$ n_1 = \frac{g_v g_s}{(2\pi)^2} \int\int d^2k \quad f(E-|e|V_+)|\psi_1|^2-f(E+|e|V_+)(1-|\psi_1|^2) =\frac{2}{\pi} \int_0^{1/\sqrt{3}a}dk\quad k \left(f(E-|e|V_+)|\psi_1|^2-f(E+|e|V_+)(1-|\psi_1|^2)\right)$$

where we've use the fact that the probability of finding a hole in the top layer is the same as not finding an electron (hence $|\psi_t^{(hole)}|^2 = 1 - |\psi_t|^2$). For $n_2$, we simply set $1\to 2$.

Our calculation will only require $n_- = n_1-n_2$, which we obtain by simply computing

$$n_- = \frac{4}{(2\pi)^2}2\pi \int_0^{1/\sqrt{3}a}dk\quad k [f(E-|e|V_+)\left(|\psi_1(k;V_-)|^2-|\psi_2(k;V_-)|^2\right) - f(E+|e|V_+)(|\psi_2|^2-|\psi_1|^2)]$$
$$\quad =\frac{2}{\pi} \int_0^{1/\sqrt{3}a}dk\quad k [f(E-|e|V_+)+f(E+|e|V_+)]\left(|\psi_1(k;V_-)|^2-|\psi_2(k;V_-)|^2\right)$$

VERSION 2: Electrons also occupy the valence band. We must consider their contribution. In general

$$n_- = \frac{4}{(2\pi)^2}2\pi \int_0^{1/\sqrt{3}a}dk\quad k \left[f(\epsilon_c-|e|V_+)\Delta|\psi|_c^2+ f(\epsilon_v-|e|V_+)\Delta|\psi|_v^2 \right]$$

where the subscripts $c$ and $v$ refer to the conduction and valence bands respectively. Recall $\epsilon_v = -\epsilon_c$. In addition, $\Delta|\psi|_c^2$ has a factor of $1/\epsilon_c$ out front. So $\Delta|\psi|_v^2 = -\Delta|\psi|_c^2$. Converting all quantities to conduction band, and dropping all subscripts,

$$n_- = \frac{4}{(2\pi)^2}2\pi \int_0^{1/\sqrt{3}a}dk\quad k \left[f(\epsilon-|e|V_+)\Delta|\psi|^2 - f(-\epsilon-|e|V_+)\Delta|\psi|^2 \right]$$

$$n_- = \frac{2}{\pi} \int_0^{1/\sqrt{3}a}dk\quad k \left[f(\epsilon-|e|V_+) - f(-\epsilon-|e|V_+)\right]\Delta|\psi|^2$$

Recall that the difference in probability $|\psi_1|^2-|\psi_2|^2$ is given by a function in `BLG.BandStructure` called `Pdiff`. This is used in defining the function `nminus`.

### Comparison of Young and Levitov

Below, we compare to an expression provided by Young and Levitov. The first equality is the expression they derive, and the second is the same quantity in our own notation

$$n_- = -\nu_0v_-\ln{\left( \frac{2\Lambda}{|v_+|+\sqrt{v_+^2-v_-^2}} \right)} = -\left(\frac{2m}{\pi\hbar^2}\right) (-|e|V_-)\ln{\left( \frac{2\epsilon(k_c)/|e|}{|V_+|+\sqrt{V_+^2-V_-^2}} \right)}$$
$$n_- = -\nu_0v_-\ln{\left( \frac{2\Lambda}{|v_-|}\right)} = -\left(\frac{2m}{\pi\hbar^2}\right) (-|e|V_-)\ln{\left( \frac{2\epsilon(k_c)/|e|}{|V_-|} \right)}$$

$\Lambda = \epsilon(k_c)/|e|$ is the cutoff energy, $\nu_0=2m/(\pi\hbar^2)$.

```python
def nminusYL(vplus,vminus):
    '''Low energy approximation for carrier density difference by Young and Levitov'''
    # Define proportionality factors out front
    nu0 = 2 * m * q / (pi * hbar**2)
    prop = nu0 * vminus
    
    # Find cutoff energy. Approximate it as the vminus=0 case
    Lambda = Dispersion( 1 / (np.sqrt(3) * a), 0, 1 ) / q
    
    # Compute the denominator of the log
    metal = abs(vplus) >= abs(vminus)
    den = (metal) * np.abs(vplus) + np.sqrt( metal * vplus**2 + (-1)**metal * vminus**2 ) 
    
    return prop * np.log(2 * Lambda / den)
```

Below is a comparison of our two expressions. It should be noted that the expression of Young and Levitov is only valid for low Fermi levels and low gaps $|e|V_-, |e|V_+ << \gamma_1$.

```python
vminus = np.linspace(-0.1,0.1, num=200)
vplus  = np.array([0.05])

nminus_array = np.empty((len(vplus),len(vminus)))
nminusYL_array = np.copy(nminus_array)

fig = plt.figure()
ax  = fig.add_axes([0,0,1,1])
ax.set_xlabel('vminus')

for i, vp in enumerate(vplus):
    nminus_array[i,:] = nminus(vp,vminus,0)
    nminusYL_array[i,:] = nminusYL(vp,vminus)
    ax.plot(vminus, nminus_array[i],label="n2 {} V".format(vp))
    ax.plot(vminus, nminusYL_array[i],label="nYL {} V".format(vp))

ax.legend(title = 'Fermi Level (V)');
```

```python
import Data_Manipulation as dm
```

```python
# Load the data
vplus = dm.get_vplus(0)
vminus = dm.get_vminus(0)

VPx , VMy = np.meshgrid(vplus,vminus)

nplus_surface = dm.get_nplus(0)
nminus_surface = dm.get_nminus(0)
```

```python
# Plot the surfaces (only every 10 points)
fig = plt.figure(figsize=(16,6))
ax1 = fig.add_subplot(121,projection='3d')
ax1.plot_surface(VPx, VMy, -q*nplus_surface, rcount=40, ccount=40, 
                cmap = 'viridis', edgecolor='none')
ax1.set_ylabel('Vminus (V)')
ax1.set_xlabel('Vplus (V)')

ax1.set_title('Total Charge Density (C/m^2)')
ax1.view_init(30,-80);

ax2 = fig.add_subplot(122,projection='3d')
ax2.plot_surface(VPx, VMy, -q*nminus_surface, rcount=40, ccount=40, 
                cmap = 'viridis', edgecolor='none')
ax2.set_ylabel('Vminus (V)')
ax2.set_xlabel('Vplus (V)')

ax2.set_title('Interlayer Charge Density Difference (C/m^2)')
ax2.view_init(30,30);
```
